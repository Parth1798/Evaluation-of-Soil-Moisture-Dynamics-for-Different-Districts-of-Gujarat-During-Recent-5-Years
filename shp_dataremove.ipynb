```python
import geopandas as gpd

# Replace 'your_shapefile.shp' with the path to your shapefile
shapefile_path ="C:\\Users\\parth\\OneDrive\\Desktop\\Parth makwana\\Moisture_project\\IndiaDistrict.shp"
gdf = gpd.read_file(shapefile_path)
```

```python
gdf
```

```python
gdf.drop(['ID_0', 'ISO', 'NAME_0', 'ID_1', 'ID_2', 'HASC_2',
       'CCN_2', 'CCA_2', 'TYPE_2', 'ENGTYPE_2', 'NL_NAME_2', 'VARNAME_2'],axis=1, inplace = True)
```

```python
field_name = "NAME_1"
for i in gdf.NAME_1:
    if i == 'Gujarat':
        continue    
    else:
        value_to_remove = i
    filtered_gdf = gdf[gdf[field_name] != value_to_remove]
    gdf = filtered_gdf
        
```

```python
gdf.index=list(range(len(gdf.index)))
```

```python
gdf.to_file("C:\\Users\\parth\\OneDrive\\Desktop\\Parth makwana\\Moisture_project\\District.shp")
```
